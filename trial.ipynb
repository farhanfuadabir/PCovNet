```python
l_win = 3
l_seq = 2
data = np.arange(0,40).astype(float)
data = np.expand_dims(data, axis=-1)
print(data, data.shape)
```

```python
dataset = []
for i in range(len(data) - (l_win * l_seq) + 1):
    dataset.append(
        data[i: i+(l_win * l_seq)])
dataset = np.array(dataset)
my = dataset.reshape(dataset.shape[0], l_seq, l_win, 1)
```

```python
print(my.shape)
# lstm_seq = np.expand_dims(lstm_seq, axis=-1)
display(my[1])
print(my[0].shape)
```

```python
for k in range(l_win):
    n_not_overlap_wins = (len(data) - k) // l_win
    n_train_lstm = n_not_overlap_wins - l_seq + 1
    cur_lstm_seq = np.zeros((n_train_lstm, l_seq, l_win))
    for i in range(n_train_lstm):
        cur_seq = np.zeros((l_seq, l_win))
        for j in range(l_seq):
          # print(k,i,j)
          cur_seq[j] = np.squeeze(data[k + l_win* (j+i): k + l_win * (j + i + 1)], axis=-1)
        cur_lstm_seq[i] = cur_seq
    if k == 0:
        lstm_seq = cur_lstm_seq
    else:
        lstm_seq = np.concatenate((lstm_seq, cur_lstm_seq), axis=0)
lstm_seq = np.expand_dims(lstm_seq, axis=-1)
```

```python
print(lstm_seq.shape)
# lstm_seq = np.expand_dims(lstm_seq, axis=-1)
display(lstm_seq[1])
print(lstm_seq[0].shape)
```
